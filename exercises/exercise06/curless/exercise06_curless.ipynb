```python
#filter df.DataFrame to only records with available bikes (new dataset of just available bikes)
df_filter = df[(df['is_reserved'] == 0) & (df['is_disabled'] == 0)]

#create additional columns within filtered data for the purpose of getting the data into hours instead of minutes
df_filter['date'] = pd.to_datetime(df_filter['timestamp']).dt.date
df_filter['hour'] = pd.to_datetime(df_filter['timestamp']).dt.hour

#remove duplicates so bikes are only counted once
df_dup = df_filter[['bike_id','date','hour']].drop_duplicates()

#start making a graphic
import seaborn as sb
df_chart = pd.DataFrame(df_dup.groupby('hour')['bike_id'].count())
sb.barplot(data = df_chart,x = 'hour',y = 'bike_id')
```
